```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Healthcare Data Analysis Project

## Introduction
Healthcare data is crucial in understanding patient demographics, medical conditions, and billing trends. This project aims to clean and analyze a healthcare dataset, providing valuable insights into patient distributions, medical conditions, and hospital billing. The results can help healthcare providers and insurers optimize services and costs.
## Data Preprocessing
Before diving into analysis, we must clean and preprocess the data to ensure accuracy and consistency.

```python
# Load Dataset
# We load the dataset to analyze patient demographics, medical conditions, and billing information
df = pd.read_csv("healthcare_dataset.csv")
```

```python
# Display basic info
# Checking dataset structure, column types, and missing values
df.info()
df.head()
```

```python
# Standardizing Column Names
# Convert column names to lowercase and replace spaces with underscores for easier access
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
# Convert Dates
# Ensuring that date columns are in the correct datetime format for better analysis
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
```

```python
# Standardize Text Columns
# Cleaning text data to ensure uniformity in values
df['gender'] = df['gender'].str.capitalize()
df['name'] = df['name'].str.title()
df['doctor'] = df['doctor'].str.title()
df['hospital'] = df['hospital'].str.title()
```

## Exploratory Data Analysis (EDA)
After preprocessing, we explore the dataset to identify patterns and trends.

### Patient Demographics Analysis
Understanding the gender and blood type distribution helps in assessing healthcare services and predicting specific medical needs.

```python
def plot_distribution(column, title):
    """ Function to visualize the distribution of categorical data."""
    plt.figure(figsize=(10, 5))
    sns.countplot(x=df[column], palette='viridis')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Understanding gender distribution among patients
plot_distribution('gender', 'Gender Distribution')

# Understanding blood type prevalence in the dataset
plot_distribution('blood_type', 'Blood Type Distribution')

# Identifying the most common medical conditions
plot_distribution('medical_condition', 'Common Medical Conditions')
```

### Billing Analysis
Financial aspects in healthcare are important for cost management and hospital efficiency. We analyze billing amounts based on admission types to understand cost variations.

```python
# Exploring billing trends based on admission type
plt.figure(figsize=(12, 6))
sns.boxplot(x='admission_type', y='billing_amount', data=df, palette='coolwarm')
plt.title('Billing Amount by Admission Type')
plt.show()
```
